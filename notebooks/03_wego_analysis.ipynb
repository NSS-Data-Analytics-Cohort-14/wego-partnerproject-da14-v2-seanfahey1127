1) This section begins parsing out the time and date points of TSP activation in the data into 
individual dataframes.

```python
#Sectioning out data from before TSP activation in dataset for route in question.
# TSP began on 2/3/2025 at 12:00pm (noon)
before_tsp = wego[(wego['CALENDAR_ID'] <= 120250203) & 
(wego['TSP_PHASE_NAME'].str.contains('Pre-TSP'))]
before_tsp.info()
```

```python
before_tsp
```

```python
##filtering dates that are included in unconditional TSP, Feb 3rd - Feb 10th, 2025 11:59am
tsp_uncon_feb = wego[(wego['CALENDAR_ID'] >= 120250203) & 
(wego['CALENDAR_ID'] <= 120250210) &
(wego['TSP_PHASE_NAME'].str.contains('TSP Unconditional 1'))]
tsp_uncon_feb.info()
```

```python
tsp_uncon_feb
```

```python
##filtering dates that are included in conditional TSP, Feb 10th - Apr 28, 2025
tsp_cond_feb = wego[(wego['CALENDAR_ID'] >= 120250210) & 
(wego['CALENDAR_ID'] <= 120250428) &
(wego['TSP_PHASE_NAME'].str.contains('TSP Schedule-Conditional'))]
tsp_cond_feb.info()
```

```python
tsp_cond_feb
```

```python
#filtering dates with TSP off between Apr 28th and May 5th
tsp_off_apr = wego[(wego['CALENDAR_ID'] >= 120250428) & 
(wego['CALENDAR_ID'] <= 120250505) &
(wego['TSP_PHASE_NAME'].str.contains('TSP Off'))]
tsp_off_apr.info()
```

```python
tsp_off_apr
```

```python
#Filtering dates for headway priority condition for TSP beginning May 12th
tsp_cond_may = wego[(wego['CALENDAR_ID'] >= 120250512) &
(wego['TSP_PHASE_NAME'].str.contains('TSP Headway-Conditional'))]
tsp_cond_may.info()
```

```python
tsp_cond_may
```
